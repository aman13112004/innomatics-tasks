# Mini Student Management System using Python Lists and Dictionaries

This notebook demonstrates how Python data structures like lists and dictionaries can be used to build a simple student management system.

Operations included:
- Add student
- Display students
- Search student
- Update marks
- Delete student

```python

students = []

print("Student database initialized")
```

```python
def add_student(name, age, marks):
    student = {
        "name": name,
        "age": age,
        "marks": marks
    }
    students.append(student)
    print("Student added successfully!")
```

```python
def display_students():
    if len(students) == 0:
        print("No student records found.")
    else:
        for s in students:
            print(s)
```

```python
def search_student(name):
    for s in students:
        if s["name"] == name:
            print("Student found:", s)
            return
    print("Student not found.")
```

```python
def update_marks(name, new_marks):
    for s in students:
        if s["name"] == name:
            s["marks"] = new_marks
            print("Marks updated successfully!")
            return
    print("Student not found.")
```

```python
def delete_student(name):
    for s in students:
        if s["name"] == name:
            students.remove(s)
            print("Student deleted successfully!")
            return
    print("Student not found.")
```

```python
# Add students
add_student("Aman", 21, 85)
add_student("Rahul", 22, 90)

# Display students
display_students()

# Search student
search_student("Aman")

# Update marks
update_marks("Aman", 95)

# Display after update
display_students()

# Delete student
delete_student("Rahul")

# Final display
display_students()
```
