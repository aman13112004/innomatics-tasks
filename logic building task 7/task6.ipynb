Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
🌍 Real-Time Use
Smart home automation systems
📥 Sample Input
Light Name: Bedroom Light
Action: ON
📤 Expected Output
Bedroom Light is ON

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default state

    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Taking input from user
light_name = input("Enter Light Name: ")
inputaction = input("Enter Action (ON/OFF): ")

light = SmartLight(light_name)

if inputaction.upper() == "ON":
    light.turn_on()
elif inputaction.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action")
```

Problem 2: Employee ID Card System
 Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.
Real-Time Use
Corporate HR systems
📥 Sample Input
Employee Name: Rahul
Employee ID: EMP102
Department: IT

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\n----- Employee ID Card -----")
        print(f"Employee Name : {self.name}")
        print(f"Employee ID   : {self.emp_id}")
        print(f"Department    : {self.department}")
        print("----------------------------")


# Taking input from user
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Creating Employee object
employee = Employee(name, emp_id, department)

# Displaying ID card
employee.display_id_card()
```

Problem 3: Mobile Contact Record
Problem Statement
Create a class to store contact name and phone number and display contact information.
Real-Time Use
Mobile contact applications
📥 Sample Input
Contact Name: Anita
Phone Number: 9876543210
📤 Expected Output
Contact Saved
Name: Anita
Phone: 9876543210

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)

contact.display_contact()
```

Problem 4: Product Price Tag Generator
Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
 Real-Time Use
Retail billing systems
📥 Sample Input
Product Name: Headphones
Price: 2499
📤 Expected Output
Product: Headphones
Price: ₹2499

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


# 📥 Taking input from user
name = input("Product Name: ")
price = input("Price: ")

# Creating Product object
product = Product(name, price)

# Printing price tag
product.print_price_tag()
```

Problem 5: Movie Rating Display System
Problem Statement
Create a class that stores movie name and rating and displays movie details.
Real-Time Use
Streaming platforms
📥 Sample Input
Movie Name: Inception
Rating: 4.8


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print("\n----- Movie Details -----")
        print(f"Movie Name : {self.name}")
        print(f"Rating     : {self.rating}")
        print("--------------------------")


# Taking input
name = input("Movie Name: ")
rating = float(input("Rating (0-5): "))

# Validation
if 0 <= rating <= 5:
    movie = Movie(name, rating)
    movie.display_details()
else:
    print("Invalid Rating! Please enter a value between 0 and 5.")
```

Problem 6: Delivery Address Manager
Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.
Real-Time Use
Courier and logistics platforms
📥 Sample Input
Customer Name: Suman
Address: Hyderabad
📤 Expected Output
Delivery Details
Customer: Suman
Address: Hyderabad

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("\nDelivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


name = input("Customer Name: ")
address = input("Address: ")


delivery = Delivery(name, address)

delivery.display_delivery_details()
```
